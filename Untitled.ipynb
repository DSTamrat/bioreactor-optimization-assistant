```python
import pandas as pd

df = pd.read_csv("data/bioreactor_synthetic_1000rows.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="viridis", font_scale=1.2)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="time_hr", y="vcd_e6_per_ml", hue="batch_id", linewidth=2.5)

plt.title("VCD Growth Curve Over Time", fontsize=18)
plt.xlabel("Time (hours)")
plt.ylabel("VCD (million cells/mL)")
plt.legend(title="Batch ID")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,6))

sns.lineplot(data=df, x="time_hr", y="glucose_g_per_L", ax=ax[0], color="blue")
ax[0].set_title("Glucose Consumption Over Time")
ax[0].set_xlabel("Time (hours)")
ax[0].set_ylabel("Glucose (g/L)")

sns.lineplot(data=df, x="time_hr", y="lactate_g_per_L", ax=ax[1], color="red")
ax[1].set_title("Lactate Accumulation Over Time")
ax[1].set_xlabel("Time (hours)")
ax[1].set_ylabel("Lactate (g/L)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="time_hr", y="do_pct", color="teal", linewidth=2)
sns.lineplot(data=df, x="time_hr", y="ph", color="orange", linewidth=2)

plt.title("DO and pH Trends Over Time")
plt.xlabel("Time (hours)")
plt.ylabel("Value")
plt.legend(["Dissolved Oxygen (%)", "pH"])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="time_hr", y="feed_rate_ml_per_hr", color="purple", linewidth=2)

plt.title("Feed Rate Profile")
plt.xlabel("Time (hours)")
plt.ylabel("Feed Rate (mL/hr)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("data/bioreactor_synthetic_1000rows.csv")

# Beautiful theme
sns.set_theme(style="whitegrid", palette="viridis", font_scale=1.2)

# Create dashboard layout
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# --- 1. VCD Growth Curve ---
sns.lineplot(data=df, x="time_hr", y="vcd_e6_per_ml", hue="batch_id", ax=axes[0,0], linewidth=2.5)
axes[0,0].set_title("VCD Growth Curve")
axes[0,0].set_xlabel("Time (hr)")
axes[0,0].set_ylabel("VCD (million cells/mL)")

# --- 2. Glucose Consumption ---
sns.lineplot(data=df, x="time_hr", y="glucose_g_per_L", ax=axes[0,1], color="blue", linewidth=2.5)
axes[0,1].set_title("Glucose Consumption")
axes[0,1].set_xlabel("Time (hr)")
axes[0,1].set_ylabel("Glucose (g/L)")

# --- 3. Dissolved Oxygen ---
sns.lineplot(data=df, x="time_hr", y="do_pct", ax=axes[1,0], color="teal", linewidth=2.5)
axes[1,0].set_title("Dissolved Oxygen (%)")
axes[1,0].set_xlabel("Time (hr)")
axes[1,0].set_ylabel("DO (%)")

# --- 4. Feed Rate Profile ---
sns.lineplot(data=df, x="time_hr", y="feed_rate_ml_per_hr", ax=axes[1,1], color="purple", linewidth=2.5)
axes[1,1].set_title("Feed Rate (mL/hr)")
axes[1,1].set_xlabel("Time (hr)")
axes[1,1].set_ylabel("Feed Rate (mL/hr)")

plt.tight_layout()
plt.show()
```
